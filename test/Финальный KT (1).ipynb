```python
%%writefile test_login.py
import pytest

# Простая серверная логика для проверки логина/пароля
def check_login(email, password):
    correct_email = "test@example.com"
    correct_password = "123456"
    return email == correct_email and password == correct_password


def test_correct_credentials():
    assert check_login("test@example.com", "123456")

def test_wrong_email():
    assert not check_login("wrong@example.com", "123456")

def test_wrong_password():
    assert not check_login("test@example.com", "wrong")

def test_both_wrong():
    assert not check_login("x@x.com", "000000")

def test_empty_email():
    assert not check_login("", "123456")

def test_empty_password():
    assert not check_login("test@example.com", "")

def test_both_empty():
    assert not check_login("", "")

def test_email_case_sensitive():
    assert not check_login("Test@example.com", "123456")  # другой регистр

def test_password_case_sensitive():
    assert not check_login("test@example.com", "123456 ")  # пробел

def test_similar_email():
    assert not check_login("test@exampl.com", "123456")  # опечатка
```
